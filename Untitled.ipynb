```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv(r"C:\Users\Soham Nagvekar\tmdb-movie-metadata\tmdb_5000_movies.csv")
credits = pd.read_csv(r"C:\Users\Soham Nagvekar\tmdb-movie-metadata\tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies_df = movies.merge(credits, on='title')
```

```python
movies_df.head(1)
```

```python
movies_df=movies_df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_df.head()
```

```python
movies_df.dropna(inplace=True)
```

```python
movies_df.isnull().sum()
```

```python
import ast
def convert(obj):
     genre_list=[]
     for i in ast.literal_eval(obj):
         genre_list.append(i['name'])
     return genre_list    
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
movies_df.head()
```

```python
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
def convert_cast(obj):
     actor_list=[]
     count=0
     for i in ast.literal_eval(obj):
         if count!=3:
             actor_list.append(i['name'])
             count+=1
         else:
            break
     return actor_list    
```

```python
movies_df['cast']= movies_df['cast'].apply(convert_cast)
```

```python
def fetch_dir(obj):
     director_list=[]
     for i in ast.literal_eval(obj):
         if i['job'] == 'Director':
             director_list.append(i['name'])
     return director_list    
```

```python
movies_df['crew']= movies_df['crew'].apply(fetch_dir)
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df
```

```python
movies_df['tags'] = movies_df['overview']+movies_df['genres']+movies_df['cast']+movies_df['crew']+movies_df['keywords']
```

```python
movies_final = movies_df.drop(['overview','genres','cast','crew','keywords'], axis=1)
```

```python
movies_final.head()
```

```python
movies_final['tags']=movies_final['tags'].apply(lambda x:" ".join(x))
```

```python
movies_final
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies_final['tags']).toarray()
```

```python
vectors
```

```python
# !pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    list=[]
    for i in text.split():
        list.append(ps.stem(i))
    return " ".join(list)  
```

```python
movies_final['tags']=movies_final['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies_final[movies_final['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_final.iloc[i[0]].title)
    return
```

```python
recommend('The Dark Knight')
```
